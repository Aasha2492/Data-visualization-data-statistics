# Titanic Data Visualization & Basic Statistics

Assignment Submission

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
# Load dataset
df = pd.read_csv("titanic_dataset.csv")
df.head()
```

## Step 1: Survival Count Bar Chart

```
df['Survived'].value_counts().plot(kind='bar')
plt.title("Survived vs Not Survived")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Number of Passengers")
plt.show()
```

## Step 2: Fare Distribution by Pclass

```
sns.boxplot(x='Pclass', y='Fare', data=df)
plt.title("Fare Distribution by Passenger Class")
plt.show()
```

## Step 3: Age Statistics

```
mean_age = df['Age'].mean()
median_age = df['Age'].median()

print("Mean Age:", mean_age)
print("Median Age:", median_age)
```

## Step 4: Learning

From the bar chart, we observe that more passengers did not survive compared to those who survived. 
The box plot shows that first-class passengers generally paid higher fares than second and third class. 
The mean and median age values help us understand the central tendency of passengers' ages.
